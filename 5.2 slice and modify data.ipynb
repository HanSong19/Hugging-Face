## Parallelization
use num_proc argument


```python
slow_tokenizer = AutoTokenizer.from_pretrained("bert-base-cased", use_fast = False)

def slow_tokenizer_function(data):
  return slow_tokenizer(data['review'], truncation = True)

tokenized_dataset = drug_dataset.map(slow_tokenizer_function, batched = True, num_proc = 8)
```

 Note, however, that won’t always be the case — for values of num_proc other than 8, our tests showed that it was faster to use batched=True without that option. In general, we don’t recommend using Python multiprocessing for fast tokenizers with batched=True.


```python
def tokenize_and_split(data):
  return tokenizer(
      data["review"],
      truncation=True,
      max_length=128,
      return_overflowing_tokens=True,
  )
```

```python
def tokenize_and_split(data):
  result = tokenizer(
      data["review"],
      truncation=True,
      max_length=128,
      return_overflowing_tokens = True
  )

  sample_map = result.pop("overflow_to_sample_mapping")
  for key, values in data.items():
      result[key] = [values[i] for i in sample_map]
  return result
```

```python
tokenized_dataset = drug_dataset.map(tokenize_and_split, batched=True)
tokenized_dataset
```

```python
train_df = drug_dataset['train'][:]
```

```python
frequencies = (
    train_df["condition"]
    .value_counts()
    .to_frame()
    .reset_index()
    .rename(columns={"index": "condition", "condition": "frequency"})
)

frequencies.head()
```

```python
rating = (
    train_df.groupby('drugName')["rating"]
    .mean()
    .to_frame()
    .reset_index()
    .rename(columns={"index": "drugname", "condition": "mean"})
)

rating.head()
```
